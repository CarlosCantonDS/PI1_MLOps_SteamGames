```python
import pandas as pd
from textblob import TextBlob
```

En este notebook se va a trabajar con el analisis de sentimientos de las reviews, <br>
para poder dar un valor numerico que este relacionado con la review de cada usuario. <br>
Debe tomar el valor '0' si es malo, '1' si es neutral y '2' si es positivo. <br>
Para cumplir en terminos de practicidad y considerando que lo que se solicita en el <br>
proyecto es un MVP, se eligió una manera simple de hacer el analisis de sentimiento <br>
para poder administrar el tiempo para otras tareas también demandantes, en este sentido <br>
Se tomó la decisión de encontrar una librería de python que use lenguaje natural y que nos <br>
ayude a resolver la tarea, esa librería es TextBlob.

```python
df_reviews = pd.read_csv('data/user_reviews_clean.csv', encoding='utf-8')
df_reviews.head()
```

```python
# Función para analizar la polaridad de una revisión
def analizar_polaridad(review):
    
    analysis = TextBlob(review)
    polaridad = analysis.sentiment.polarity
    if polaridad < -0.1:
        return 0  # Revisión mala
    elif polaridad > 0.1:
        return 2  # Revisión positiva
    else:
        return 1  # Revisión neutra
```

```python
# Aplicar la función de análisis de polaridad a las revisiones
df_reviews['sentiment_analysis'] = df_reviews['reviews_review'].apply(analizar_polaridad)
```

```python
df_reviews.head()
```

```python
# Se procede a eliminar la columna reviews_reviews
df_reviews = df_reviews.drop(columns=['reviews_review'])
df_reviews.columns
```

```python
reviews_clean = 'data/user_reviews_clean.csv'
df_reviews.to_csv(reviews_clean, index=False, encoding='utf-8')
```
