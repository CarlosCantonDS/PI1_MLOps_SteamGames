```python
import pandas as pd
import json
%load_ext autoreload
%autoreload 2
```

```python
steam_games_ruta = 'data/output_steam_games.json'

# El json se va a leer linea por linea 
filas_json = []
with open(steam_games_ruta) as f:
    for line in f.readlines():
        data = json.loads(line)
        filas_json.append(data)

# Se almacena en un data frame de pandas
df_games = pd.DataFrame(filas_json)
df_games
```

```python
df_games.isna().sum()
```

Desde un inicio se observa que hay muchos valores nulos,<br>
y todas las columnas tienen un valor similar de nulos; se procede a eliminarlos

```python
df_games = df_games.dropna().reset_index(drop=True)
df_games.isna().sum()
```

```python
df_games.info()
```

En busca de optimizar y conocer el data set se revisan algunas columnas <br>
Con base en los outputs anteriores 

```python
# se observa tipo de dato y valores unicos de early_access
print(df_games['early_access'].dtype)
print("---"*10)
df_games['early_access'].unique()
```

```python
# se cambia el tipo de dato de string a bool
df_games['early_access'] = df_games['early_access'].astype(bool)
```

```python
#   se observa tipo de dato y valores unicos de price
print(df_games['price'].dtype)
print("---"*10)
df_games['price'].unique()
```

```python
# se cambia el tipo de dato de string a flotante y se asigna cero para los casos especiales
df_games['price'] = df_games['price'].apply(lambda x: round(float(x), 3) if str(x).replace('.', '', 1).isdigit() else 0).astype('float32')
```

```python
#   se observa tipo de dato y valores unicos de price
print(df_games['price'].dtype)
print("---"*10)
df_games['price'].unique()
```

Se revisa la columna generos para posteriormente tomar la decision <br>
de hacer explode de las columnas y luego hacer One-Hot-Encoding

```python
#   se observa tipo de dato y valores unicos de genres
print(df_games['genres'].dtype)
print("---"*10)
df_games['genres'].unique
```

```python
df_games = df_games.explode('genres')
df_games = df_games.dropna(subset=['genres'])
df_games
```

```python
df_games.shape
```

Se procede a eliminar las columnas que no son necesarias para la API <br>
siguiendo la información de los endpoints solicitados, más adelante en <br>
el modelo de Recomendacion de videojuegos se entrará en detalle  <br>
la eleccion de las columnas para el algoritmo

```python
df_games = df_games.drop(['specs', 'url', 'reviews_url','tags'], axis=1)
df_games.head(4)
```

Siguiendo con la limpieza y optimizacion se toma la decisión <br>
de cambiar el release date por el año, no es necesario especificar <br>
mes y día 

```python
#   se observa tipo de dato y valores unicos de price
print(df_games['release_date'].dtype)
print("---"*10)
df_games['release_date'].unique()
```

```python
# Asumiendo que la columna 'release_date' es del formato 'yyyy-mm-dd'
df_games['release_year'] = df_games['release_date'].apply(lambda x: x.split('-')[0])

# Ahora, 'release_year' contiene solo el año y se elimina la columna 'release_date'
df_games.drop(columns=['release_date'], inplace=True)
```

```python
df_games['release_year'].unique()
```

```python
# Define una función lambda para extraer el año si el valor contiene las primeras tres letras de un mes
def extract_year(value):
    try:
        # Si el valor se puede dividir en partes y la última parte es un año, se devuelve
        parts = value.split()
        last_part = parts[-1]
        if last_part.isdigit() and len(last_part) == 4:
            return last_part
    except:
        pass
    return pd.NA

# Reemplaza "SOON" y "SOON™" con "Not-Released-Yet"
df_games['release_year'] = df_games['release_year'].replace(['SOON', 'SOON™'], 'Not-Released-Yet')

# Aplica la función lambda a la columna 'release_year'
df_games['release_year'] = df_games['release_year'].apply(extract_year)
```

```python
ruta_df_clean = 'data/steam_games_clean.csv'
df_games.to_csv(ruta_df_clean, index=False, encoding='utf-8')
```
